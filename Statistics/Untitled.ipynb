```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos desde el archivo CSV
encuesta_df = pd.read_csv('encuesta_cdmx.csv')
```

```python
# Visualizar las primeras filas para entender la estructura de los datos
print(encuesta_df.head())
```

```python
sns.set_style("whitegrid")
```

```python
# Graficar la distribución de edades
plt.figure(figsize=(10, 6))
sns.histplot(data=encuesta_df, x='¿En qué rango de edad te encuentras?', bins=10, kde=True, color='skyblue', alpha=0.7)
plt.title('Distribución de Edades')
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Guardar la gráfica como una imagen JPG
plt.savefig('distribucion_edades.jpg', format='jpg')
```

```python
encuesta_df['¿En qué rango de edad te encuentras?'].value_counts().reset_index()
```

```python
# Graficar la distribución de género con color skyblue y cantidades
plt.figure(figsize=(8, 6))
sns.countplot(x='¿Cuál es tu género?', data=encuesta_df, palette=['skyblue', 'skyblue'])
plt.title('Distribución de Género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')

# Agregar cantidades a cada barra
for index, value in enumerate(encuesta_df['¿Cuál es tu género?'].value_counts()):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Guardar la gráfica como una imagen JPG
plt.savefig('distribucion_genero.jpg', format='jpg')
```

```python
encuesta_df['¿Cuál es tu género?'].value_counts().reset_index()
```

```python
# Graficar el tiempo manejando en CDMX con color skyblue y cantidades
plt.figure(figsize=(10, 6))
sns.countplot(x='¿Cuánto tiempo llevas manejando dentro de la Ciudad de México?', data=encuesta_df, palette=['skyblue', 'skyblue'])
plt.title('Tiempo manejando en CDMX')
plt.xlabel('Años manejando en CDMX')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')

# Agregar cantidades a cada barra
for index, value in enumerate(encuesta_df['¿Cuánto tiempo llevas manejando dentro de la Ciudad de México?'].value_counts()):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Guardar la gráfica como una imagen JPG
plt.savefig('distribucion_tiempo_manejando.jpg', format='jpg')
```

```python
encuesta_df['¿Cuánto tiempo llevas manejando dentro de la Ciudad de México?'].value_counts().reset_index()
```

```python
# Graficar la frecuencia de uso de un chatbot de tránsito 
plt.figure(figsize=(8, 6))
sns.barplot(y='Si existiera un chatbot sencillo y rápido que responda preguntas de tránsito, ¿Con qué frecuencia lo llegarías a usar?', 
              data=encuesta_df['Si existiera un chatbot sencillo y rápido que responda preguntas de tránsito, ¿Con qué frecuencia lo llegarías a usar?'].value_counts().reset_index(), x='index',  palette=['skyblue', 'skyblue'])
plt.title('Frecuencia de uso de un chatbot de tránsito')
plt.xlabel('Frecuencia de uso')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')

# Agregar cantidades a cada barra
for index, value in enumerate(encuesta_df['Si existiera un chatbot sencillo y rápido que responda preguntas de tránsito, ¿Con qué frecuencia lo llegarías a usar?'].value_counts()):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Guardar la gráfica como una imagen JPG
plt.savefig('distribucion_uso_chatbot_transito.jpg', format='jpg')
```

```python
encuesta_df['Si existiera un chatbot sencillo y rápido que responda preguntas de tránsito, ¿Con qué frecuencia lo llegarías a usar?'].value_counts().reset_index()
```

```python
# Graficar la disposición hacia un chatbot sobre el sistema legal mexicano
plt.figure(figsize=(8, 6))
sns.countplot(x='¿Te interesaría un chatbot que respondiera preguntas no solo relacionadas con el tránsito, sino con todo el sistema legal mexicano?', 
              data=encuesta_df, palette=['skyblue', 'skyblue'])
plt.title('Disposición hacia un chatbot sobre el sistema legal mexicano')
plt.xlabel('Interés')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')

# Agregar cantidades a cada barra
for index, value in enumerate(encuesta_df['¿Te interesaría un chatbot que respondiera preguntas no solo relacionadas con el tránsito, sino con todo el sistema legal mexicano?'].value_counts()):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Guardar la gráfica como una imagen JPG
plt.savefig('distribucion_chatbot_legal.jpg', format='jpg')
```
